## **1. 데이터 불러오기**

```python
URL = "https://jehyunlee.github.io/2021/10/05/Python-DS-85-popkr/"
df_popkrM = pd.read_pickle(URL + "df_popkrMT.pkl")   # 남성 인구 데이터
df_popkrF = pd.read_pickle(URL + "df_popkrFT.pkl")   # 여성 인구 데이터
   
print("# 남성 인구 데이터")
display(df_popkrM.head())
print("\n# 여성 인구 데이터")
display(df_popkrF.head())
```

## **3. 데이터가 그려질 공간 준비** : Figure, Axes

```python
fig, axs = plt.subplots(figsize=(10, 5),          # 공간의 크기는 10 x 5 inch
                        ncols=2,                  # 좌우 두 개로 나누어 Axes 둘 생성.
                        sharey=True,              # 두 Axes의 y 좌표를 공유
                        gridspec_kw={"wspace":0}) # Axes 사이 거리 = 0 # 두개의 스페이스를 붙여줌
```

## **4. 연령대별 인구 표현: 수평방향 bar plot**

```python
fig, axs = plt.subplots(ncols=2, sharey=True, figsize=(10, 5), gridspec_kw={"wspace":0})

# 가로로 눕히고 시픙ㄴ 경우엔 bar에 h 붙이면 됨, 안붙이면 세로로 나옴
axs[0].barh(df_popkrM["나이"], df_popkrM["전국"], color="green")      # 왼쪽 Axes: 남성
axs[1].barh(df_popkrF["나이"], df_popkrF["전국"], color="darkorange") # 오른쪽 Axes: 여성
```

## **5. 남성 데이터 좌우 반전**

```python
fig, axs = plt.subplots(ncols=2, sharey=True, figsize=(10, 5), gridspec_kw={"wspace":0})
axs[0].barh(df_popkrM["나이"], df_popkrM["전국"], color="green")
axs[1].barh(df_popkrF["나이"], df_popkrF["전국"], color="darkorange")

xmax = 4.5e6              # 데이터를 담기에 충분히 큰 값 (450만)
axs[0].set_xlim(xmax, 0)  # 왼쪽 남성 데이터 (큰 값에서 작은 값으로)
axs[1].set_xlim(0, xmax)  # 오른쪽 여성 데이터 ( 작은 값에서 큰 값으로 생성)
```

## **6. 눈금 레이블 수정**

```python
fig, axs = plt.subplots(ncols=2, sharey=True, figsize=(10, 5), gridspec_kw={"wspace":0})
axs[0].barh(df_popkrM["나이"], df_popkrM["전국"], color="green")
axs[1].barh(df_popkrF["나이"], df_popkrF["전국"], color="darkorange")

xmax = 4.5e6
axs[0].set_xlim(xmax, 0)
axs[1].set_xlim(0, xmax)

# 인식하기 편하도록 눈금에 for 루프를 돌려서 백만 단위를 표시해준다
xticks = np.arange(0, xmax, 1e6)          # 0부터 백만 단위로 x축 눈금 지정
for ax in axs:                            # 두 Axes에 같은 작업 적용
    ax.set_xticks(xticks)                                   # xticks 적용
    ax.set_xticklabels([f"{int(x*1e-6)}백만"                # xticklabels 적용
                        if x != 0 else "0" for x in xticks])
```

## **7. 제목, 배경 눈금 추가**

```python
fig, axs = plt.subplots(ncols=2, sharey=True, figsize=(10, 5), gridspec_kw={"wspace":0})
axs[0].barh(df_popkrM["나이"], df_popkrM["전국"], color="green")
axs[1].barh(df_popkrF["나이"], df_popkrF["전국"], color="darkorange")

xmax = 4.5e6
axs[0].set_xlim(xmax, 0)
axs[1].set_xlim(0, xmax)

xticks = np.arange(0, xmax, 1e6)
for ax, title in zip(axs, ["남성", "여성"]):
    ax.set_xticks(xticks)
    ax.set_xticklabels([f"{int(x*1e-6)}백만" 
                        if x != 0 else "0" for x in xticks])
    ax.set_title(title, color="gray", fontweight="bold", pad=16)  # 제목
    ax.grid(color="lightgray")                                    # 배경 눈금 - 그리드 그려줌, 색 지정 가능
```

```python
fig, axs = plt.subplots(ncols=2, sharey=True, figsize=(10, 5), gridspec_kw={"wspace":0})
axs[0].barh(df_popkrM["나이"], df_popkrM["전국"], color="green")
axs[1].barh(df_popkrF["나이"], df_popkrF["전국"], color="darkorange")

xmax = 4.5e6
axs[0].set_xlim(xmax, 0)
axs[1].set_xlim(0, xmax)

xticks = np.arange(0, xmax, 1e6)
for ax, title in zip(axs, ["남성", "여성"]):
    ax.set_xticks(xticks)
    ax.set_xticklabels([f"{int(x*1e-6)}백만" 
                        if x != 0 else "0" for x in xticks])
    ax.set_title(title, color="gray", fontweight="bold", pad=16)  
    ax.grid(color="lightgray")            

for ax in axs:                                       # text 삽입을 위한 loop
    for i, p in enumerate(ax.patches):               # 막대 순회 - 막대 하나하나에 접근한다. (enumerate이므로 몇번째인지에 대한 순서가 i로 온다)
        w = p.get_width()                            # 너비에서 데이터 추출
        ax.text(w, i, f" {format(w, ',')} ",         # 천단위 쉼표 추가, text삽입
                fontsize="small", va="center", ha="right")     

      # 현재 트렌드적인 측면에서 남성 여성 컬러를 중성적인 컬러로 많이 표현함. 참고 (빨강파랑 x)                   
```

```python
fig, axs = plt.subplots(ncols=2, sharey=True, figsize=(10, 5), gridspec_kw={"wspace":0})
axs[0].barh(df_popkrM["나이"], df_popkrM["전국"], color="green")
axs[1].barh(df_popkrF["나이"], df_popkrF["전국"], color="darkorange")

xmax = 6e6                            # 숫자까지 담기도록 범위 키우기 !!
axs[0].set_xlim(xmax, 0)
axs[1].set_xlim(0, xmax)

xticks = np.arange(0, xmax, 1e6)
for ax, title in zip(axs, ["남성", "여성"]):
    ax.set_xticks(xticks)
    ax.set_xticklabels([f"{int(x*1e-6)}백만" 
                        if x != 0 else "0" for x in xticks])
    ax.set_title(title, color="gray", fontweight="bold", pad=16)  
    ax.grid(color="lightgray")            

for ax in axs:                                       
    for i, p in enumerate(ax.patches):               
        w = p.get_width()                 #조건문을 걸어서 데이터 표기를 맞춰준다           
        if ax == axs[0]:              # 남성 데이터
            ha = "right"              # 글자 오른쪽에 막대
        else:                         # 여성 데이터
            ha = "left"               # 글자 왼쪽에 막대
        ax.text(w, i, f" {format(w, ',')} ",         
                fontsize="small", va="center", ha=ha)    
```

## **9. 산책 한번 하고 오기**

image.png



*** 고칠 점 발견!**

1.   글자와 눈금이 겹쳐서 지저분함  
2.   어느 지역의 통계인지 알 수 없음

```python
fig, axs = plt.subplots(ncols=2, sharey=True, figsize=(10, 5), gridspec_kw={"wspace":0})

c_M = "green"                      # 남성 데이터 색상 변수 지정
c_F = "darkorange"                 # 여성 데이터 색상 변수 지정
axs[0].barh(df_popkrM["나이"], df_popkrM["전국"], color=c_M) # 색상 변수 사용
axs[1].barh(df_popkrF["나이"], df_popkrF["전국"], color=c_F) # 색상 변수 사용

xmax = 6e6                            
axs[0].set_xlim(xmax, 0)
axs[1].set_xlim(0, xmax)

xticks = np.arange(0, xmax, 1e6)
for ax, title in zip(axs, ["남성", "여성"]):
    ax.set_xticks(xticks)
    ax.set_xticklabels([f"{int(x*1e-6)}백만" 
                        if x != 0 else "0" for x in xticks])
    ax.set_title(title, color="gray", fontweight="bold", pad=16)   # set title은 axis 하나에 붙는 타이틀
    ax.grid(axis="x", color="lightgray")                     # 세로방향 grid만 사용

for ax in axs:                                       
    for i, p in enumerate(ax.patches):               
        w = p.get_width()                            
        if ax == axs[0]:              
            ha = "right"
            c = c_M                                          # 색상 변수 사용
        else:                         
            ha = "left"               
            c = c_F                                          # 색상 변수 사용
        ax.text(w, i, f" {format(w, ',')} ", 
                c=c, alpha=0.5,                              # 색상 변수 사용, 반투명
                fontsize="small", va="center", ha=ha) 
        
        #전체에 붙여주는 타이틀은 suptitle
fig.suptitle("                 전국",  
             fontsize="medium", fontweight="bold")         # 전체 제목
fig.tight_layout()                                           # 구성요소간 간격 조정
```
